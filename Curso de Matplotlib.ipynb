# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,9,10)
x
```

```python
y = x**2 # x elevado a 2
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'red')
```

```python
plt.plot(x,y,'green')
```

```python
plt.plot(x,y,'purple')
```

```python
plt.plot(x,y,'red')
plt.title("titulo: este es un grfico")
plt.xlabel("nombre eje x")
plt.ylabel("nombre eje y")
```

```python
plt.subplot(1,2,1) #(n fila, n columna, posicin)
plt.plot(x,y,'g')
plt.subplot(1,2,2)
plt.plot(y,x,'blue')
```

```python
plt.subplot(2,1,1) #(n fila, n columna, posicin)
plt.plot(x,y,'g')
plt.subplot(2,1,2)
plt.plot(y,x,'blue')
```

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.8,0.8])
grafico.plot(x,y,'g')
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x,y,'r')

grafico2 = figura.add_axes([0.3,0.4,0.3,0.3])
grafico2.plot(y, x,'b')
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9]) 
#lista de valores izquierda, abajo, ancho, alto del grfico
grafico1.plot(x,y,'r')

grafico2 = figura.add_axes([0.3,0.5,0.3,0.3])
grafico2.plot(y, x,'b')
```

# Multigrficos (filas,columnas)

```python
figura, graficos = plt.subplots(nrows=4,ncols=3)
```

```python
figura, graficos = plt.subplots(nrows=2,ncols=2)
```

```python
figura, graficos = plt.subplots(nrows=2,ncols=2)
graficos[0][0].plot(x,y,'b')
graficos[0][1].plot(x,y,'r')
graficos[1][0].plot(x,y,'y')
graficos[1][1].plot(x,y,'g')

graficos[0][0].set_title('azul')
graficos[0][1].set_title('rojo')
graficos[1][0].set_title('amarillo')
graficos[1][1].set_title('verde')
```

# Tamao del grfico

```python
figura = plt.figure(figsize=(8,4))
grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'r')

figura = plt.figure(figsize=(10,2))
grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'r')
```

# Dos grficos dentro de una misma figura

```python
figura = plt.figure()
graficos = figura.add_axes([0.1,0.1,0.9,0.9])
graficos.plot(x,x**2,label="al cuadradro")
graficos.plot(x,x**3,label="al cubo")
```

```python
figura = plt.figure()
graficos = figura.add_axes([0.1,0.1,0.9,0.9])
graficos.plot(x,x**2,label="al cuadradro")
graficos.plot(x,x**3,label="al cubo")
graficos.legend(loc=0)
```

# Color del grfico

```python
figura = plt.figure()
graficos = figura.add_axes([0,0,1,1])
graficos.plot(x,y,'turquoise')
```

```python
figura = plt.figure()
graficos = figura.add_axes([0,0,1,1])
graficos.plot(x,y,'#a4ccee')
```

# Lnea del grfico (grosor, transparencia y tipo)

```python
figura = plt.figure()
graficos = figura.add_axes([0,0,1,1])
graficos.plot(x,y,'r',linewidth=10, alpha=0.25, linestyle=':') 
#Grosor lnea, 
#transparencia (0,1 ms transparente y 0.9 menos trasparente),
#tipo de lnea('-' ':' '-.' '--')
```
